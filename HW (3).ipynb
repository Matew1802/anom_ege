```python
import matplotlib.pyplot as plt
from math import sqrt
import random


f = open("0.txt")
s = f.readline()
a = [list(map(float, i.replace(',', '.').split())) for i in f if i.strip()]

cl = []
anom = []
for i in range(len(a)):
    if a[i] != '*':
        b = []
        b.append(a[i])
        a[i] = '*'
        for l in b:
            x, y = l
            for j in range(i+1 , len(a)):
                if a[j] != '*':
                    if ((x - a[j][0])**2 + (y-a[j][1])**2)**0.5 <= 1:
                        b.append(a[j])
                        a[j] = '*'
        if len(b) >= 30:
            cl.append(b)
        else:
            for k in range(len(b)):
                anom.append(b[k])

sm_x = sm_y = 0
for k in range(len(cl)):
    s_x = s_y = 0
    mn = 10000000000000000
    for j in cl[k]:
        x, y = j
        d = 0
        for i in cl[k]:
            d += ((x- i[0])**2+(y-i[1])**2)**0.5
        if d < mn:
            mn = d
            s_x, s_y = x, y
    sm_x += s_x
    sm_y += s_y
print((sm_x/(len(cl))*100000))
print((sm_y/(len(cl))*100000))
colors = []
for _ in range(len(cl)):
    color = f'#{random.randint(0, 0xFFFFFF):06x}'
    while color == '000000':
        color = f'#{random.randint(0, 0xFFFFFF):06x}'
    colors.append(color)

plt.figure(figsize=(8, 6))

for i, cluster in enumerate(cl):
    x, y = zip(*cluster)
    plt.scatter(x, y, color=colors[i], label=f'Кластер {i+1}')

if anom:
    x_anom, y_anom = zip(*anom)
    plt.scatter(x_anom, y_anom, color='black', label='Аномалии')

plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Распределение точек')
plt.grid(True)

plt.show()
```
