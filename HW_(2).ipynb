```python

import random

f = open("0.txt")
s = f.readline()
a = [list(map(float, i.replace(',', '.').split())) for i in f]

#цикл для нахождения любой любой точки кластера
#for i in  range(len(a)):
#    if a[i][0] > 2 and a[i][0] < 3 and a[i][1] < -1:
#        print(i)
#        break

cl = [[a[14]], [a[1]]]
a[14] = a[1] = '*'

for k in range(2):
    for i in cl[k]:
        x, y = i
        for j in range(len(a)):
            if a[j] != '*':
                if ((x - a[j][0])**2 + (y - a[j][1])**2)**0.5 <= 1:
                    cl[k].append(a[j])
                    a[j] = '*'

sm_x = sm_y = 0

for k in cl:
    s_x = s_y = 0
    mn = 10000000000000000
    for j in k:
        x, y = j
        d = 0
        for i in k:
            d += ((x- i[0])**2+(y-i[1])**2)**0.5
        if d < mn:
            mn = d
            s_x, s_y = x, y
    sm_x += s_x
    sm_y += s_y
print((sm_x/2*100000))
print((sm_y/2*100000))

colors = []
for _ in range(len(cl)):
    color = f'#{random.randint(0, 0xFFFFFF):06x}'
    while color == '000000':
        color = f'#{random.randint(0, 0xFFFFFF):06x}'
    colors.append(color)

plt.figure(figsize=(8, 6))

for i, cluster in enumerate(cl):
    x, y = zip(*cluster)
    plt.scatter(x, y, color=colors[i], label=f'Кластер {i+1}')

anom = [p for p in a if p != '*']
if anom:
    x_anom, y_anom = zip(*anom)
    plt.scatter(x_anom, y_anom, color='black', label='Аномалии')

plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Распределение точек')
plt.grid(True)

plt.show()
```
